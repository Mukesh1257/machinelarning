```python
# Max power
#first we have to remove bhp character type from data so that its convertable into floats

def fxn(x):
    return x.split()[0]
df['max_power']=df['max_power'].apply(fxn)
df
#converting into float
#df['max_power']=['max_power'].astype("float32")#Error

#find out illegal value that is not convertable into float
index=[]
for i in range(len(df)):
    try:
        float(df.iloc[i,-2])
    except:
        index.append(i)
print(index)
#Drop the illegal value index
df=df.drop(index=index)
df=df.reset_index(drop=True)

#Again convert cplumn into float
df['max_power']=df['max_power'].astype("float32")
df
```

```python
#engine
def fxn(x):
    return x.split()[0]
df['engine']=df['engine'].apply(fxn)
df
#converting into float
#df['engine']=['engine'].astype("float32")#Error

#find out illegal value that is not convertable into float
index=[]
for i in range(len(df)):
    try:
        float(df.iloc[i,-3])
    except:
        index.append(i)
print(index)
#Drop the illegal value index
df=df.drop(index=index)
df=df.reset_index(drop=True)


#Again convert cplumn into float
df['engine']=df['engine'].astype("float32")
df
```

```python
#Mileage
def fxn(x):
    return x.split()[0]
df['mileage']=df['mileage'].apply(fxn)
df['mileage']=df['mileage'].astype("float32")
df
 
```

```python
#when are data is normally distibuted-->method used for outlier detection
#Z score

#mean=10 std=2
#-2 10 +2
#8 and 12-->belong to first std deviation
#6 and 14-->belong to second std deviation
#4 and 16--->belong to third std deviation
#if any data point falling out of data range will be consider as outlier

#according to gaussian
#if the data is normally distributed,then 68% of data-->first std dev
#95% of data -->second std dev
#99.7% of data-->third std dev
```

```python
# -4 19 +4
#Z score
#(x-mean)/std
#if a value has a z score of more than +3 or less than -3 then it will be considered as an outlier
#z score returns standard deviation of each data points wrt to mean......
```

```python
def detect(column):
    mean=column.mean()
    std=column.std()
    z_score=(column-mean)/std
    return column[(z_score>3)|(z_score<-3)]
#out=detect(numeric["km_driven"])
outlier=detect(numeric["km_driven"])
print(outlier)


f=numeric['km_driven'].isin(outlier)
numeric=numeric[~f]
category=category[~f]

f=numeric['mileage'].isin(outlier)
numeric=numeric[~f]
category=category[~f]

f=numeric['max_power'].isin(outlier)
numeric=numeric[~f]
category=category[~f]

numeric=numeric.reset_index(drop=True)
category=category.reset_index(drop=True)
```
